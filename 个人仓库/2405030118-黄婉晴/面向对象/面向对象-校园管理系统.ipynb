```python
#创建学校类
class School:
    def __init__(self,name,address):
        self.name = name
        self.address = address
        self.staff_list = []               #聚合关系（学校包含员工，可独立存在
        self.motto = "自强不息，厚德载物"  # 组合关系（校训，不能独立于学校存在
    
    def count_staff_num(self):
        return len(self.staff_list)
        
    def new_staff_enrollment(self,staff_obj):
        self.staff_list.append(staff_obj.name)
        print(f"[{staff_obj.name}]已成为学校员工...")  
        
#创建分校类       
class BranchSchool(School):                 #继承School
    def __init__(self,name,address,headquarter):
        super().__init__(name,address)
        self.headquarter = headquarter     #关联关系（每个分校关联一个总校对象

#创建班级类
class Class:
    def __init__(self,class_num,course_obj,school_obj):
        self.class_num = class_num
        self.course_obj = course_obj       #关联关系（每个班级关联一个课程对象
        self.school_obj = school_obj       #关联关系（每个班级关联一个学校对象
        self.students = []                 #聚合关系（班级包含学生
        
    def create_teaching_record(self):
        print(f"[{self.course_obj}]课程记录中...")
        
    def add_student(self,student_obj):
        self.students.append(student_obj)
        print(f"[{student_obj.name}]加入了班级[{self.class_num}]")
        
    def drop_out(self,student_obj):
        if student_obj in self.students:
            self.students.remove(student_obj)
            print(f"[{student_obj}]退学了...")
        else:
            print(f"[{student_obj}]不在班级中，退学操作失败...")

#创建课程类
class Course:
    def __init__(self,name,price,outline):
        self.name = name
        self.price = price
        self.outline = outline

#创建员工类
class Staff:
    def __init__(self,name,age,position):
        self.name = name
        self.age = age
        self.position = position
        self.salary = self.calculate_salary
    
    def calculate_salary(self):
        position_salary = {'教授':10000,'讲师':6000,'助教':4000,'厨师':4000,'保洁':3500}
        return position_salary.get(self.position,0)
            
    def get_salary(self):
        return self.salary

#创建教师类
class Teacher(Staff):                            #继承Staff
    def __init__(self,name,age,position):
        super().__init__(name,age,position)
        self.teaching_class = []                 #一对多关联（一个教师可关联多个班级对象
    
    def teaching(self,class_obj):                #依赖class对象
        self.teaching_class.append(class_obj)
        print(f"[{self.name}]老师教[{class_obj.class_num}]的课程")

        

#创建学生类
class Student:
    def __init__(self,name,age,degree,class_obj,balance):
        self.name = name
        self.age = age
        self.degree = degree
        self.class_obj = class_obj                #关联关系（每个学生关联一个班级对象
        self.balance = balance
        
    def pay_tuition(self,course_obj):             #依赖关系（依赖course对象
        if self.balance >= course_obj.price:
            self.balance -= course_obj.price
            print(f"支付成功！[{self.name}]的账户剩余[{self.balance}]元")
        else:
            print(f"支付失败，[{self.name}]的账户余额不足")
```

```python
#实例化
def main():
    school1 = School("惠州学院","惠城区")
    branch_school1 = BranchSchool("惠州高中","河南岸校区",school1)
    course_Math = Course("数学分析",500,"极限定理，微积分...")
    course_English = Course("英语语法",300,"从句运用")
    class1 = Class("24级大一1班","course_Math","school1")
    class2 = Class("24级高一2班","course_English","branch_school1") 
    
    teacher1 = Teacher("张三","45","教授")
    teacher2 = Teacher("李四","25","助教")
    
    school1.new_staff_enrollment(teacher1)
    school1.new_staff_enrollment(teacher2)
    
    teacher1.calculate_salary()
    teacher2.calculate_salary()
    
    student1 = Student("小明","18","大一","class1",5000)
    student2 = Student("小红","16","高一","class2",3000)
    
    class1.add_student(student1)
    class2.add_student(student2)
    
    student1.pay_tuition(course_Math)
    student2.pay_tuition(course_English)
    
    teacher1.teaching(class1)
    

main()
```
