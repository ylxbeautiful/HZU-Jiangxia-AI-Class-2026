```python
#张量算法的基本性质
A = torch.arange(20,dtype=torch.float32).reshape(5,4)
B = A.clone()  #通过分配新内存，将A的一个副本分配给B
A , A+B        #具有相同形状的任意两个张量，任何按元素二元运算的结果都将是相同形状的张量
```

```python
A * B  #逐元素乘法（两个矩阵的按元素乘法称为Hadamard积
```
