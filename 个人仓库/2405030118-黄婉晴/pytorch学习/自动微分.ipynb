```python
import torch
```

```python
x = torch.arange(4.0)
x
```

```python
#在计算y对x梯度之前，需要一个地方存储梯度
x.requires_grad_(True)  #等价于在创建x时加上requires_grad=True
x.grad
```

```python
y = 2*torch.dot(x,x) #2||x||2 导数为2*2x
y
```

```python
#调用反向传播函数来自动计算y关于x每个分量的梯度
y.backward()
x.grad
```

```python
x.grad == 4*x
```

```python
# 在默认情况下，PyTorch会累积梯度，我们需要清除之前的值
x.grad.zero_()
y = x.sum()
y.backward()
x.grad
```

```python
#向量y关于向量x的导数
x.grad.zero_()
y=x*x
y.sum().backward()
x.grad
```

```python
#分离计算,分离y来返回一个新变量u,将u作为常数处理
x.grad.zero_()
y = x*x
u = y.detach()
z = u*x

z.sum().backward()
x.grad == u
```

```python
x.grad.zero_()
y.sum().backward()
x.grad==2*x
```

```python
def f(a):
    b = a*2
    while b.norm() <1000:
        b = b*2
    if b.sum()>0:
        c = b
    else:
        c = 100*b
    return c
```

```python
a = torch.randn(size=(), requires_grad=True)
d = f(a)
d.backward()
```

```python
a.grad == d / a
```
