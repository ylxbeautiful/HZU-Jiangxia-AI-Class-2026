# 填充和步幅

填充
* 在卷积神经网络中，填充（padding）是一种在输入图像边缘添加额外的像素值的方法，以控制输出特征图的大小。填充可以增加特征图的宽度，从而增加卷积层的感受野。填充还可以帮助保持输入和输出特征图的空间维度一致。

```python
from mxnet import nd
from mxnet.gluon import nn

def comp_conv2d(conv2d,X):
    conv2d.initialize()
    X = X.reshape((1,1)+X.shape) # 增加两个维度，第一个是批量大小，第二个是通道数
    Y = conv2d(X)
    return Y.reshape(Y.shape[2:]) # 排除前两个维度（批量大小和通道数）
```

```python
conv2d=nn.Conv2D(1,kernel_size=3,padding=1) # 两侧分别填充1行或列
X=nd.random.uniform(shape=(8,8))
comp_conv2d(conv2d,X).shape
```

```python
con2d=nn.Conv2D(1,kernel_size=(5,3),padding=(2,1))
comp_conv2d(conv2d,X).shape
```

步幅
* 步幅（stride）是卷积操作中卷积核在输入图像上移动的步长。步幅决定了卷积核在输入图像上移动的速度，从而影响输出特征图的大小。较小的步幅会导致输出特征图更大，而较大的步幅会导致输出特征图更小。

* 填充可以增加输出的高和宽，这常用于保持输入和输出特征图的空间维度一致。
* 步幅可以减小输出的高和宽，这常用于加速计算或减少参数数量，例如输出的高和宽为输入的高和宽的1/n。
