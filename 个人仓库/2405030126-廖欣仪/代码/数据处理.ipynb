```python
#创建人工数据集，存储在csv文件
data_file=r"D:\house_tiny.csv"#绝对路径
with open(data_file,'w')as f:
    f.write('NumRooms,Alley,Price\n')#列名
    f.write('NA,Pave,127500\n')
    f.write('2,NA,106000\n')
    f.write('4,NA,178100\n')
    f.write('NA,NA,140000\n')
print("文件以创建")
```

```python
#加载原始数据
import pandas as pd
data=pd.read_csv(data_file)
print(data)
```

```python
#处理缺失数据
#插值
inputs,outputs=data.iloc[:,:2],data.iloc[:,2]#取出前两列为inputs，最后一列为outputs
inputs['NumRooms']=inputs['NumRooms'].fillna(inputs['NumRooms'].mean())#将非空平均值填入空值
inputs=pd.get_dummies(inputs,dummy_na=True).astype(int)#将布尔类型转为0、1
print(inputs)
```
