```python
#元素全0
torch.zeros((2,3,4))
```

```python
#元素全1
torch.ones((2,3,4))
```

```python
#特定赋值
torch.tensor([[2,1,4,3],[1,2,3,4],[4,3,2,1]])#二维
```

```python
#标准算术运算符,元素进行
a=torch.tensor([1.0,2,4,8])#浮点
b=torch.tensor([2,2,2,2])#整数
a+b,a-b,a*b,a/b,a**b#加，减，乘，除，平方
```

```python
#cat连结
X=torch.arange(12,dtype=torch.float32).reshape((3,4))#指定浮点
Y=torch.tensor([[2.0,1,4,3],[1,2,3,4],[4,3,2,1]])
torch.cat((X,Y),dim=0),torch.cat((X,Y),dim=1) #0行，1列
```

```python
#构建二维张量
X==Y
```

```python
#对所有元素进行求和
X.sum()
```

```python
#广播机制
c=torch.arange(3).reshape((3,1))#3行1列
d=torch.arange(2).reshape((1,2))#1行2列
print(c)
print(d)
c+d
```

```python
#访问元素
print(X)
X[-1]
X[1:3]
```

```python
#指定元素写入矩阵
X[1,2]=9
X
```

```python
#为新结果分配内存
before=id(Y)
Y=Y+X
id(Y)==before
```

```python
#执行原地操作
Z=torch.zeros_like(Y)
print('id(Z):',id(Z))
Z[:]=X+Y
print('id(Z):',id(Z))
```

```python
#减少操作的内存开销
before=id(X)
X+=Y
id(X)==before
```

```python
#转换为numpy张量
A=X.numpy()
B=torch.tensor(A)
type(A),type(B)
```
