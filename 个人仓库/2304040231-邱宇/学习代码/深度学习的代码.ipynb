```python

class Dog:
    def __init__(self,name,breed):
        self.name = name,
        self.breed = breed,
        self.attack = 30
        self.life_val = 100
    def bite(self,person):
        person.life_val -= self.attack
        print("person name is",person.name,"now life_val is",person.life_val)
class Person:
    def __init__(self,name,age):
        self.name = name
        self.age = age
        self.life_val = 100
        self.attack = 20
    def fight(self,dog):
        dog.life_val -= self.attack
        print("dog name is",dog.name,"now life_val is",dog.life_val)
d1 = Dog("wangcai","chuanchuan")
p1 = Person("QY",21)
d1.bite(p1)
p1.fight(d1)
```

```python
#继承的代码和思路
class Animal:
    def __init__(self,name,age,sex):
        self.name = name
        self.age = age
        self.sex = sex
        print("----父类----")
    def eat(self):
        print("吃了一些食物")
class Person(Animal):
    def __init__(self,name,age,sex,hobby):
        Animal.__init__(self,name,age,sex)
        self.hobby = hobby
    def talk(self):
        print("talk with  friend")
p1 = Person("Qy",18,"M","luguan")
p1.talk()
p1.eat()
print(p1.hobby)

#多继承
class ShenXian:
    def fly(self):
        print("shenxian can fly the sky")
    def fight(self):
        print("shenxian can use magic fight people")
class Monkey:
    def eat(self):
        print("monkey like eat peach")
    def fight(self):
        print("monkey can use hand fight people")
class MonkeyKing(ShenXian,Monkey):
    pass
mk = MonkeyKing()
mk.fly()
mk.fight()
#多继承是属于广度优先，在两个是同一个源头的情况下
```
