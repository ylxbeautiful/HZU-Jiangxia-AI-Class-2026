```python
import pandas as pd
data = pd.read_csv(r'd:\Users\44462\Desktop\house_tiny.csv')#加载数据
data
```

```python
#插值处理缺失的数据
inputs,outputs = data.iloc[:,0:2],data.iloc[:,2]
inputs['NumRooms'] = inputs['NumRooms'].fillna(inputs['NumRooms'].mean())
inputs
```
