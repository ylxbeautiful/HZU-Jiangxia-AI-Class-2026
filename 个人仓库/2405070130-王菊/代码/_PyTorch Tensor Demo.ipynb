17-2自动求导实现

```python
import torch
x = torch.arange(4.0)
x
```

```python
#计算x的梯度之前，需要一个地方存放梯度
x.requires_grad_(True)   #等价于“x = torch.arange(4.0,requires_grad=True)
x.grad  #默认为None
```

```python
y=2*torch.dot(x,x)   #计算y ,torch.dot(x, x) 表示两个一维张量的点积（内积
y
```

```python
#调用反向传播函数来自动计算y关于x每个分量的梯度
y.backward()  #求导
x.grad  #访问导数
```

```python
x.grad== 4*x  #验算求导结果是否正确
```

```python
#z在默认条件下，pytorch会累积梯度，需要手动清理
x.grad.zero_()  #将x梯度清零
#计算另外的函数
y=x.sum()
y.backward() #求导
x.grad
```

需要批量中的每个样本单独计算的偏导数之和。
对非标量调用‘backward’需要传入‘grandient’参数，该参数指定微分函数

```python
x.grad.zero_()  #将梯度清零
y = x * x
y.sum().backward() #等价于y.backward(torch.ones(len(x)))
x.grad
```

```python
#将某些计算移到计算图之外
x.grad.zero_()  #清零梯度
y=x*x
u=y.detach()  # 分离y得到u（u脱离计算图，视为常数）
z=u*x

z.sum().backward()   # 反向传播，计算z求和后对x的梯度
x.grad == u     #验证梯度结果
```

```python
x.grad.zero_()  #清零梯度
y.sum().backward()  # 反向传播，计算y求和后对x的梯度
x.grad==2*x
```

```python
#计算图需要复杂的python控制流的时候，依旧可以求导
def f(a):
    b = a *2
    while b.norm() < 1000:      # 循环：当b的范数小于1000时，持续将b翻倍
        b=b * 2
    if b.sum() >0:
        c=b
    else:
        c=100* b
    return c
a=torch.randn(size=(),requires_grad=True)
d=f(a)  # 前向传播计算d=f(a)
d.backward()  # 反向传播求d对a的梯度
a.grad == d/a  # 验证梯度：a.grad等于d/a（因d=k*a，导数为k=d/a）
```
